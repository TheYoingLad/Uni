```python
# hsv = hue saturation value
# [0-167] [0-255] [0-255]

hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
```

```python
mask = cv.inRange(hsv,np.array([[90,50,70]]),np.array([[128,255,255]]))
print(mask)
```

```python
highlighted = cv.bitwise_and(img, img, mask=mask)

cv.imshow("title",highlighted)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

cv.imshow("title",gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
edges = cv.Canny(gray, 100, 200)

plt.subplot(121),plt.imshow(gray, cmap="gray")
plt.title("original image"),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(edges, cmap="gray")
plt.title("edges"),plt.xticks([]),plt.yticks([])
```

```python
img_denoise = cv.fastNlMeansDenoising(img, None, 10, 7, 21)

plt.subplot(121),plt.imshow(img)
plt.subplot(122),plt.imshow(img_denoise)
plt.show()
```

```python
img5 = cv.imread("people.jpg")

if img5 is None:
    raise ValueError("does not exist")

face_cascade = cv.CascadeClassifier("haarcascades_frontalface_default.xml")

gray_people = cv.cvtColor(img5, cv.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray_people, scaleFactor=1.1, minNeighbors=5)

for (x,y,w,h) in faces:
    cv.rectangle(img5, (x,y), (x+w, y+h))

```
